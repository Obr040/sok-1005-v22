```python
import pandas as pd

import numpy as np

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rangelak/fifa-stats/master/data/fifa-world-cup.csv", index_col=0)

df
```

```python
kontinent = ['Sør-Amerika', 'Afrika', 'Europa', 'Asia', 'Europa', 'Nord-Amerika', 'Europa', 'Sør-Amerika', 'Europa', 'Sør-Amerika', 'Europa', 'Sør-Amerika', 'Europa', 'Sør-Amerika', 'Europa', 'Europa', 'Sør-Amerika', 'Europa', 'Europa', 'Sør-Amerika']

df['CONTINENT'] = kontinent

COLORS ={'Afrika': '#E14827', 'Sør-Amerika': '#84E127', 'Asia': '#2792E1', 'Europa': '#BC27E1', 'Nord-Amerika': '#E04A6C'}

df['COLORS']=df['CONTINENT'].apply(lambda cont: COLORS[cont])

df
```

```python
df['SIZE']=df['AVERAGE_GOALS'].apply(lambda pop: pop**2.5/0.5)
df[df['CONTINENT']=='Sør-Amerika']

from bokeh.plotting import figure

p = figure(
        title = "VM i fotball og gjennomsnittlige mål scoret", 
        x_axis_label = 'År',
        y_axis_label = 'Antall mål',
    
        x_axis_type = "log",
    
        tools="hover", 
        tooltips = [
            ("Arrangør","@LOCATION"),
            ("År","@YEAR"),
            ("Vinner","@WINNER"),
            ("Antall kamper","@MATCHES"),
            ("Antall lag","@TEAMS"),
            ("Antall mål","@GOALS"),
            ("Gjennomsnittlige tilskuere","@AVERAGE_ATTENDANCE")
            ],
    
        plot_height = 750,
        plot_width = 1500)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file

p.scatter(
    source=df,
    x= 'YEAR', 
    y= 'GOALS',
    
    size='SIZE', 
    color= 'COLORS',
    legend_field='CONTINENT',
    alpha= 0.8
    )
p.legend.location = "top_left"
```

```python
output_notebook()
output_file("VM.html")
show(p)
```
